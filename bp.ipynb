```python
def reduce_rare_categories(df, column, threshold=20):
  
    top_categories = df[column].value_counts().nlargest(threshold).index
    
    # Replace categories not in the top N with 'Other'
    return df[column].where(df[column].isin(top_categories), 100)

# Apply this function to columns with high cardinality
high_cardinality_columns = ['alertid', 'incidentid', 'accountupn', 'accountname', 'ipaddress','accountsid','accountobjectid','alerttitle','filename','devicename','url','sha256','folderpath','deviceid','detectorid','orgid']  # You can modify this list
for col in high_cardinality_columns:
    df[col] = reduce_rare_categories(df, col, threshold=20)

# Now df contains 'Other' for rare categories
df.head()
```
